```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

import warnings
warnings.filterwarnings('ignore')
```

# 🔢 Data Prepping & Cleaning

### ✅ Import Data

```python
# Load the CSV file into a DataFrame
stores = pd.read_csv("stores.csv")
products = pd.read_csv("products.csv")
orders = pd.read_csv("orders.csv")
customers = pd.read_csv("customers.csv", encoding='ISO-8859-1')
```

## ✅ Clean & Wrangle

```python
#update all column names for easier coding
for df in [stores, products, orders, customers]:
    df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
#split remove color from product_name
products['product_name'] = products.apply(lambda row: row['product_name'].split(row['color'])[0].strip(), axis=1)
```

```python
# format relevant columns
orders['order_date'] = pd.to_datetime(orders['order_date'])
products['unit_price_usd'] = products['unit_price_usd'].replace('[\$,]', '', regex=True).astype(float)
products['unit_cost_usd'] = products['unit_cost_usd'].replace('[\$,]', '', regex=True).astype(float)
```

```python
order_details = orders.merge(customers, on = 'customerkey', how='left')
order_details = order_details.merge(stores, on = 'storekey', how='left')
order_details = order_details.merge(products, on = 'productkey', how='left')

order_details['total_revenue_usd'] = order_details['unit_price_usd'] * order_details['quantity']
order_details['total_costs_usd'] = order_details['unit_cost_usd'] * order_details['quantity'] #technically COGS
order_details['total_gross_profit_usd'] = order_details['total_revenue_usd'] - order_details['total_costs_usd']

# rename and remove duplicate column that were a result of the above joins
order_details = order_details.loc[:, ~order_details.columns.str.contains("_y")]
order_details.columns = order_details.columns.str.replace('_x', '', regex=False)
#print(order_details.info())
order_details['order_year'] = order_details['order_date'].dt.year
order_details.head()
```

```python
customer_transactions = (
    order_details
    .groupby(['customerkey', 'order_number', 'order_date', 'order_year'])[['total_revenue_usd', 'total_gross_profit_usd']]
    .sum()
    .reset_index()
    .rename(columns={
        'customerkey': 'customer_id',
        'order_number': 'order_id'
    })
)
customer_transactions.head()
```

```python
customer_summary = customer_transactions.groupby("customer_id",as_index = False).agg(
      acquisition_date=("order_date", "min"),
      acquisition_year=("order_year", "min"),
      first_order_date=("order_date", "min"),
      last_order_date=("order_date", "max"),
      total_orders=("order_id", "nunique"),
      total_revenue_usd=("total_revenue_usd", "sum"),
      total_profit_usd=("total_gross_profit_usd", "sum"),
      ).reset_index()


# Simulate acquisition channel and CAC -------------------------

np.random.seed(42)

# Sort customers by total profit so higher-value customers tend to skew better channels
customer_summary = customer_summary.sort_values("total_profit_usd").reset_index(drop=True)

# Create customer percentile rank
customer_summary["customer_percentile"] = (
    (customer_summary.index + 1) / len(customer_summary)
)

# Assign acquisition channel using simple business logic
def assign_channel(p):
    if p <= 0.50:
        # Bottom 50% customers
        return np.random.choice(
            ["Paid Search", "Organic Search", "Direct"],
            p=[0.60, 0.30, 0.10]
        )
    elif p <= 0.85:
        # Middle 35% customers
        return np.random.choice(
            ["Paid Search", "Organic Search", "Direct"],
            p=[0.45, 0.35, 0.20]
        )
    else:
        # Top 15% customers
        return np.random.choice(
            ["Paid Search", "Organic Search", "Direct"],
            p=[0.25, 0.40, 0.35]
        )

customer_summary["acquisition_channel"] = customer_summary["customer_percentile"].apply(assign_channel)

# Simulate CAC based on acquisition channel
cac_ranges = {
    "Paid Search": (20, 45),
    "Organic Search": (1, 8),
    "Direct": (0, 5)
}

def simulate_cac(channel):
    low, high = cac_ranges[channel]
    return round(np.random.uniform(low, high), 2)

customer_summary["cac_usd"] = customer_summary["acquisition_channel"].apply(simulate_cac).astype(float)

# ---------------------------------------------------


customer_summary["clv_profit_usd"] = (
    customer_summary["total_profit_usd"] - customer_summary["cac_usd"]
)

customer_summary["lifetime_days"] = (
    customer_summary["last_order_date"] - customer_summary["first_order_date"]
).dt.days

customer_summary["is_repeat_customer"] = (
    customer_summary["total_orders"] > 1
).astype(int)

customer_transactions = customer_transactions.merge(customer_summary[['customer_id', 'acquisition_date','acquisition_year','acquisition_channel','cac_usd']], on='customer_id',how='inner')
customer_transactions["days_since_first_order"] = (customer_transactions["order_date"] - customer_transactions["acquisition_date"]).dt.days
customer_transactions
```

```python
# -----------------------------------
# CUSTOMER x PERIOD ACTIVITY
# one row per customer per years-since-joining period
# -----------------------------------

customer_transactions["years_since_joining"] = np.floor(customer_transactions["days_since_first_order"] / 365.25).astype(int)

customer_period_activity = (
    customer_transactions.groupby(["customer_id", "acquisition_year", "years_since_joining"])
      .agg(
          revenue_usd=("total_revenue_usd", "sum"),
          profit_usd=("total_gross_profit_usd", "sum"),
          orders=("order_id", "nunique")
      )
      .reset_index()
)

customer_period_activity["active_flag"] = 1
customer_period_activity["period_label"] = (
    "Year " + customer_period_activity["years_since_joining"].astype(str)
)

customer_period_activity.head()

```

```python
# -----------------------------------
# COHORT PERIOD SUMMARY
# one row per cohort per year-since-joining
# -----------------------------------
import plotly.express as px

# Cohort size = number of customers acquired in each cohort year
cohort_sizes = (
    customer_summary.groupby("acquisition_year")["customer_id"]
    .nunique()
    .rename("cohort_size")
    .reset_index()
)

# Cohort base revenue = revenue in Year 0 for each cohort
cohort_base_revenue = (
    customer_period_activity.loc[customer_period_activity["years_since_joining"] == 0]
    .groupby("acquisition_year")["revenue_usd"]
    .sum()
    .rename("cohort_base_revenue")
    .reset_index()
)

cohort_period_summary = (
    customer_period_activity.groupby(["acquisition_year", "years_since_joining"])
    .agg(
        customers_active=("customer_id", "nunique"),
        revenue_usd=("revenue_usd", "sum"),
        profit_usd=("profit_usd", "sum")
    )
    .reset_index()
    .merge(cohort_sizes, on="acquisition_year", how="left")
    .merge(cohort_base_revenue, on="acquisition_year", how="left")
)

cohort_period_summary["customer_retention_pct"] = (
    cohort_period_summary["customers_active"] / cohort_period_summary["cohort_size"] * 100
)

cohort_period_summary["revenue_retention_pct"] = (
    cohort_period_summary["revenue_usd"] / cohort_period_summary["cohort_base_revenue"] * 100
)

cohort_period_summary.head()
```

# PHASE 1: Revenue Diagnostics

```python
# Business Question: What changed in the retailer's revenue engine during 2020
# compared to historical sales patterns? Early 2021 is treated as directional context only.

# This script focuses on:
# 1. Cleaning the order detail data
# 2. Creating order-level data from line-item data
# 3. Building monthly and annual business performance tables
# 4. Comparing 2020 against the 2016–2019 baseline
# 5. Comparing each 2020 month against the same month historically
# 6. Analyzing category mix and gross profit changes
# 7. Creating clean output tables for charts and executive summaries
```

### Cleaning
The raw data is at the line-item level. That means one order can appear across multiple rows if the order included multiple products. Before we analyze anything, we need to make sure dates and numeric columns are in the right format.

```python
df = order_details.copy()

# Convert dates to datetime. If a value cannot be converted, it becomes NaT, which is pandas' version of a missing date.
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["delivery_date"] = pd.to_datetime(df["delivery_date"], errors="coerce")

# These are not central to the analysis, but cleaning them now keeps the dataset consistent.
df["birthday"] = pd.to_datetime(df["birthday"], errors="coerce")
df["open_date"] = pd.to_datetime(df["open_date"], errors="coerce")

# Remove rows where order_date is missing. Without an order date, the row cannot be used in time-based analysis.
df = df[df["order_date"].notna()].copy()

# Create clean time fields.
df["order_year"] = df["order_date"].dt.year
df["order_month"] = df["order_date"].dt.month
df["order_month_name"] = df["order_date"].dt.month_name()
df["order_month_start"] = df["order_date"].dt.to_period("M").dt.to_timestamp()
df["order_quarter"] = df["order_date"].dt.to_period("Q").astype(str)

# These are the numeric columns.
numeric_cols = [
    "quantity",
    "unit_cost_usd",
    "unit_price_usd",
    "total_revenue_usd",
    "total_costs_usd",
    "total_gross_profit_usd",
    "square_meters"
]

# Convert numeric fields to numbers. If a value is bad or blank, it becomes NaN.
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Replace missing financial values with 0. This helps prevent aggregation errors.
# **In a real client project, we would also profile missingness separately.
financial_cols = [
    "total_revenue_usd",
    "total_costs_usd",
    "total_gross_profit_usd"
]

df[financial_cols] = df[financial_cols].fillna(0)

# Replace missing quantity with 0.
df["quantity"] = df["quantity"].fillna(0)
```

### Define Analysis Periods
- Historical Baseline: 2016–2019
- Pandemic Year: 2020
- Early 2021: incomplete directional context

```python
def assign_analysis_period(year):
    if year in [2016, 2017, 2018, 2019]:
        return "Historical Baseline"
    elif year == 2020:
        return "Pandemic Year"
    elif year == 2021:
        return "Early 2021 Directional"
    else:
        return "Other"

df["analysis_period"] = df["order_year"].apply(assign_analysis_period)

# Keep only the years relevant to this case study. This prevents accidental inclusion of years outside the intended scope.
df = df[df["analysis_period"] != "Other"].copy()
```

```python



# ------------------------------------------------------------
# 5. CREATE AN ORDER-LEVEL TABLE
# ------------------------------------------------------------
# The source data is line-item level.
#
# Example:
# One order may have:
# - line 1: camera
# - line 2: microwave
#
# If we count orders directly from the raw line-item table,
# we could accidentally double count orders.
#
# So we create one row per order.

order_summary = (
    df.groupby(
        [
            "order_number",
            "customerkey",
            "order_date",
            "order_month_start",
            "order_year",
            "order_month",
            "analysis_period"
        ],
        as_index=False
    )
    .agg(
        order_revenue_usd=("total_revenue_usd", "sum"),
        order_costs_usd=("total_costs_usd", "sum"),
        order_gross_profit_usd=("total_gross_profit_usd", "sum"),
        order_quantity=("quantity", "sum"),
        line_item_count=("line_item", "count")
    )
)

# Calculate order-level gross margin.
# Gross margin tells us how much profit the company keeps from revenue
# before other operating expenses.
order_summary["order_gross_margin_pct"] = np.where(
    order_summary["order_revenue_usd"] > 0,
    order_summary["order_gross_profit_usd"] / order_summary["order_revenue_usd"],
    np.nan
)


# ------------------------------------------------------------
# 6. MONTHLY BUSINESS PERFORMANCE
# ------------------------------------------------------------
# This table is one row per month.
#
# It answers:
# - How much revenue did the business generate?
# - How many customers purchased?
# - How many orders were placed?
# - How much did customers spend per order?
# - How profitable were the sales?
#
# This is the main table for executive trend analysis.

monthly_performance = (
    order_summary.groupby(
        ["order_month_start", "order_year", "order_month", "analysis_period"],
        as_index=False
    )
    .agg(
        revenue_usd=("order_revenue_usd", "sum"),
        gross_profit_usd=("order_gross_profit_usd", "sum"),
        customers=("customerkey", "nunique"),
        orders=("order_number", "nunique"),
        quantity_sold=("order_quantity", "sum")
    )
)

# Average Order Value.
# This tells us how much revenue the average order generated.
monthly_performance["avg_order_value_usd"] = np.where(
    monthly_performance["orders"] > 0,
    monthly_performance["revenue_usd"] / monthly_performance["orders"],
    np.nan
)

# Orders per customer.
# This is a simple purchase frequency metric.
monthly_performance["orders_per_customer"] = np.where(
    monthly_performance["customers"] > 0,
    monthly_performance["orders"] / monthly_performance["customers"],
    np.nan
)

# Revenue per customer.
# This combines customer volume, frequency, and order value.
monthly_performance["revenue_per_customer_usd"] = np.where(
    monthly_performance["customers"] > 0,
    monthly_performance["revenue_usd"] / monthly_performance["customers"],
    np.nan
)

# Gross margin.
monthly_performance["gross_margin_pct"] = np.where(
    monthly_performance["revenue_usd"] > 0,
    monthly_performance["gross_profit_usd"] / monthly_performance["revenue_usd"],
    np.nan
)


# ------------------------------------------------------------
# 7. ANNUAL BUSINESS PERFORMANCE
# ------------------------------------------------------------
# This table is one row per year.
#
# It gives leadership a clean annual view before we go into
# month-level diagnostics.

annual_performance = (
    order_summary.groupby(
        ["order_year", "analysis_period"],
        as_index=False
    )
    .agg(
        revenue_usd=("order_revenue_usd", "sum"),
        gross_profit_usd=("order_gross_profit_usd", "sum"),
        customers=("customerkey", "nunique"),
        orders=("order_number", "nunique"),
        quantity_sold=("order_quantity", "sum")
    )
)

annual_performance["avg_order_value_usd"] = np.where(
    annual_performance["orders"] > 0,
    annual_performance["revenue_usd"] / annual_performance["orders"],
    np.nan
)

annual_performance["orders_per_customer"] = np.where(
    annual_performance["customers"] > 0,
    annual_performance["orders"] / annual_performance["customers"],
    np.nan
)

annual_performance["revenue_per_customer_usd"] = np.where(
    annual_performance["customers"] > 0,
    annual_performance["revenue_usd"] / annual_performance["customers"],
    np.nan
)

annual_performance["gross_margin_pct"] = np.where(
    annual_performance["revenue_usd"] > 0,
    annual_performance["gross_profit_usd"] / annual_performance["revenue_usd"],
    np.nan
)

annual_performance = annual_performance.sort_values("order_year").copy()


# ------------------------------------------------------------
# 8. HISTORICAL BASELINE VS 2020
# ------------------------------------------------------------
# This is the main executive comparison.
#
# We compare:
# - Average annual performance from 2016–2019
# - Actual full-year 2020 performance
#
# Why average annual baseline?
# Because comparing 2020 to only one year, like 2019, can be noisy.
# A 2016–2019 average gives a more stable historical benchmark.

baseline_years = [2016, 2017, 2018, 2019]
comparison_year = 2020

baseline_annual = (
    annual_performance[annual_performance["order_year"].isin(baseline_years)]
    .agg(
        revenue_usd=("revenue_usd", "mean"),
        gross_profit_usd=("gross_profit_usd", "mean"),
        customers=("customers", "mean"),
        orders=("orders", "mean"),
        quantity_sold=("quantity_sold", "mean"),
        avg_order_value_usd=("avg_order_value_usd", "mean"),
        orders_per_customer=("orders_per_customer", "mean"),
        revenue_per_customer_usd=("revenue_per_customer_usd", "mean"),
        gross_margin_pct=("gross_margin_pct", "mean")
    )
    .reset_index()
    .rename(columns={"index": "metric", 0: "historical_baseline_value"})
)

# The aggregation above creates a vertical table.
# Now get 2020 values in the same structure.

pandemic_2020 = (
    annual_performance[annual_performance["order_year"] == comparison_year]
    .agg(
        revenue_usd=("revenue_usd", "sum"),
        gross_profit_usd=("gross_profit_usd", "sum"),
        customers=("customers", "sum"),
        orders=("orders", "sum"),
        quantity_sold=("quantity_sold", "sum"),
        avg_order_value_usd=("avg_order_value_usd", "mean"),
        orders_per_customer=("orders_per_customer", "mean"),
        revenue_per_customer_usd=("revenue_per_customer_usd", "mean"),
        gross_margin_pct=("gross_margin_pct", "mean")
    )
    .reset_index()
    .rename(columns={"index": "metric", 0: "value_2020"})
)

baseline_vs_2020 = baseline_annual.merge(
    pandemic_2020,
    on="metric",
    how="left"
)

# Calculate absolute and percent difference.
baseline_vs_2020["absolute_change"] = (
    baseline_vs_2020["value_2020"]
    - baseline_vs_2020["historical_baseline_value"]
)

baseline_vs_2020["pct_change"] = np.where(
    baseline_vs_2020["historical_baseline_value"] != 0,
    baseline_vs_2020["absolute_change"] / baseline_vs_2020["historical_baseline_value"],
    np.nan
)


# ------------------------------------------------------------
# 9. MONTH-MATCHED BASELINE VS 2020
# ------------------------------------------------------------
# Retail is seasonal.
#
# January should be compared to historical January performance.
# November should be compared to historical November performance.
#
# This avoids misleading conclusions from comparing a strong month
# to a naturally weak month.

# First, build the 2016–2019 monthly baseline.
historical_monthly_baseline = (
    monthly_performance[monthly_performance["order_year"].isin(baseline_years)]
    .groupby("order_month", as_index=False)
    .agg(
        baseline_revenue_usd=("revenue_usd", "mean"),
        baseline_gross_profit_usd=("gross_profit_usd", "mean"),
        baseline_customers=("customers", "mean"),
        baseline_orders=("orders", "mean"),
        baseline_quantity_sold=("quantity_sold", "mean"),
        baseline_avg_order_value_usd=("avg_order_value_usd", "mean"),
        baseline_orders_per_customer=("orders_per_customer", "mean"),
        baseline_revenue_per_customer_usd=("revenue_per_customer_usd", "mean"),
        baseline_gross_margin_pct=("gross_margin_pct", "mean")
    )
)

# Now isolate 2020 monthly performance.
performance_2020_by_month = (
    monthly_performance[monthly_performance["order_year"] == 2020]
    .copy()
    .rename(columns={
        "revenue_usd": "revenue_2020_usd",
        "gross_profit_usd": "gross_profit_2020_usd",
        "customers": "customers_2020",
        "orders": "orders_2020",
        "quantity_sold": "quantity_sold_2020",
        "avg_order_value_usd": "avg_order_value_2020_usd",
        "orders_per_customer": "orders_per_customer_2020",
        "revenue_per_customer_usd": "revenue_per_customer_2020_usd",
        "gross_margin_pct": "gross_margin_2020_pct"
    })
)

# Join the historical baseline to the 2020 monthly data.
monthly_2020_vs_baseline = performance_2020_by_month.merge(
    historical_monthly_baseline,
    on="order_month",
    how="left"
)

# Add readable month name.
monthly_2020_vs_baseline["month_name"] = pd.to_datetime(
    monthly_2020_vs_baseline["order_month"].astype(str),
    format="%m"
).dt.month_name()

# Calculate differences for each key metric.
monthly_2020_vs_baseline["revenue_change_usd"] = (
    monthly_2020_vs_baseline["revenue_2020_usd"]
    - monthly_2020_vs_baseline["baseline_revenue_usd"]
)

monthly_2020_vs_baseline["revenue_pct_change"] = np.where(
    monthly_2020_vs_baseline["baseline_revenue_usd"] != 0,
    monthly_2020_vs_baseline["revenue_change_usd"]
    / monthly_2020_vs_baseline["baseline_revenue_usd"],
    np.nan
)

monthly_2020_vs_baseline["customer_change"] = (
    monthly_2020_vs_baseline["customers_2020"]
    - monthly_2020_vs_baseline["baseline_customers"]
)

monthly_2020_vs_baseline["customer_pct_change"] = np.where(
    monthly_2020_vs_baseline["baseline_customers"] != 0,
    monthly_2020_vs_baseline["customer_change"]
    / monthly_2020_vs_baseline["baseline_customers"],
    np.nan
)

monthly_2020_vs_baseline["orders_change"] = (
    monthly_2020_vs_baseline["orders_2020"]
    - monthly_2020_vs_baseline["baseline_orders"]
)

monthly_2020_vs_baseline["orders_pct_change"] = np.where(
    monthly_2020_vs_baseline["baseline_orders"] != 0,
    monthly_2020_vs_baseline["orders_change"]
    / monthly_2020_vs_baseline["baseline_orders"],
    np.nan
)

monthly_2020_vs_baseline["aov_change_usd"] = (
    monthly_2020_vs_baseline["avg_order_value_2020_usd"]
    - monthly_2020_vs_baseline["baseline_avg_order_value_usd"]
)

monthly_2020_vs_baseline["aov_pct_change"] = np.where(
    monthly_2020_vs_baseline["baseline_avg_order_value_usd"] != 0,
    monthly_2020_vs_baseline["aov_change_usd"]
    / monthly_2020_vs_baseline["baseline_avg_order_value_usd"],
    np.nan
)

monthly_2020_vs_baseline["gross_profit_change_usd"] = (
    monthly_2020_vs_baseline["gross_profit_2020_usd"]
    - monthly_2020_vs_baseline["baseline_gross_profit_usd"]
)

monthly_2020_vs_baseline["gross_profit_pct_change"] = np.where(
    monthly_2020_vs_baseline["baseline_gross_profit_usd"] != 0,
    monthly_2020_vs_baseline["gross_profit_change_usd"]
    / monthly_2020_vs_baseline["baseline_gross_profit_usd"],
    np.nan
)

monthly_2020_vs_baseline["gross_margin_point_change"] = (
    monthly_2020_vs_baseline["gross_margin_2020_pct"]
    - monthly_2020_vs_baseline["baseline_gross_margin_pct"]
)

monthly_2020_vs_baseline = monthly_2020_vs_baseline.sort_values("order_month")


# ------------------------------------------------------------
# 10. CATEGORY PERFORMANCE BY YEAR
# ------------------------------------------------------------
# This helps us understand whether 2020 revenue came from the
# same categories as prior years.
#
# This is important because total sales can hide product mix shifts.

category_annual_performance = (
    df.groupby(["order_year", "analysis_period", "category"], as_index=False)
    .agg(
        revenue_usd=("total_revenue_usd", "sum"),
        gross_profit_usd=("total_gross_profit_usd", "sum"),
        customers=("customerkey", "nunique"),
        orders=("order_number", "nunique"),
        quantity_sold=("quantity", "sum")
    )
)

category_annual_performance["gross_margin_pct"] = np.where(
    category_annual_performance["revenue_usd"] > 0,
    category_annual_performance["gross_profit_usd"] / category_annual_performance["revenue_usd"],
    np.nan
)

category_annual_performance["avg_order_value_usd"] = np.where(
    category_annual_performance["orders"] > 0,
    category_annual_performance["revenue_usd"] / category_annual_performance["orders"],
    np.nan
)


# ------------------------------------------------------------
# 11. CATEGORY MIX BY YEAR
# ------------------------------------------------------------
# Category revenue alone is useful, but revenue share is often
# more useful strategically.
#
# Example:
# Computers may still be the largest category, but if they make
# up a smaller share of revenue in 2020 than before, that matters.

yearly_revenue_totals = (
    category_annual_performance.groupby("order_year", as_index=False)
    .agg(total_company_revenue_usd=("revenue_usd", "sum"))
)

yearly_gross_profit_totals = (
    category_annual_performance.groupby("order_year", as_index=False)
    .agg(total_company_gross_profit_usd=("gross_profit_usd", "sum"))
)

category_mix_by_year = category_annual_performance.merge(
    yearly_revenue_totals,
    on="order_year",
    how="left"
).merge(
    yearly_gross_profit_totals,
    on="order_year",
    how="left"
)

category_mix_by_year["revenue_share"] = np.where(
    category_mix_by_year["total_company_revenue_usd"] > 0,
    category_mix_by_year["revenue_usd"]
    / category_mix_by_year["total_company_revenue_usd"],
    np.nan
)

category_mix_by_year["gross_profit_share"] = np.where(
    category_mix_by_year["total_company_gross_profit_usd"] > 0,
    category_mix_by_year["gross_profit_usd"]
    / category_mix_by_year["total_company_gross_profit_usd"],
    np.nan
)


# ------------------------------------------------------------
# 12. CATEGORY BASELINE VS 2020
# ------------------------------------------------------------
# This shows which categories changed most in 2020 compared to
# their 2016–2019 annual average.
#
# This helps answer:
# - Which categories drove growth?
# - Which categories drove decline?
# - Did revenue shift into different types of products?
# - Did profit shift differently than revenue?

category_baseline = (
    category_annual_performance[
        category_annual_performance["order_year"].isin(baseline_years)
    ]
    .groupby("category", as_index=False)
    .agg(
        baseline_revenue_usd=("revenue_usd", "mean"),
        baseline_gross_profit_usd=("gross_profit_usd", "mean"),
        baseline_customers=("customers", "mean"),
        baseline_orders=("orders", "mean"),
        baseline_quantity_sold=("quantity_sold", "mean"),
        baseline_gross_margin_pct=("gross_margin_pct", "mean"),
        baseline_avg_order_value_usd=("avg_order_value_usd", "mean")
    )
)

category_2020 = (
    category_annual_performance[
        category_annual_performance["order_year"] == 2020
    ]
    .drop(columns=["analysis_period"])
    .rename(columns={
        "revenue_usd": "revenue_2020_usd",
        "gross_profit_usd": "gross_profit_2020_usd",
        "customers": "customers_2020",
        "orders": "orders_2020",
        "quantity_sold": "quantity_sold_2020",
        "gross_margin_pct": "gross_margin_2020_pct",
        "avg_order_value_usd": "avg_order_value_2020_usd"
    })
)

category_2020_vs_baseline = category_baseline.merge(
    category_2020,
    on="category",
    how="outer"
)

# Fill missing numeric values with 0 for categories that may exist
# in one period but not the other.
category_numeric_cols = [
    col for col in category_2020_vs_baseline.columns
    if col != "category"
]

category_2020_vs_baseline[category_numeric_cols] = (
    category_2020_vs_baseline[category_numeric_cols].fillna(0)
)

# Revenue change.
category_2020_vs_baseline["revenue_change_usd"] = (
    category_2020_vs_baseline["revenue_2020_usd"]
    - category_2020_vs_baseline["baseline_revenue_usd"]
)

category_2020_vs_baseline["revenue_pct_change"] = np.where(
    category_2020_vs_baseline["baseline_revenue_usd"] != 0,
    category_2020_vs_baseline["revenue_change_usd"]
    / category_2020_vs_baseline["baseline_revenue_usd"],
    np.nan
)

# Gross profit change.
category_2020_vs_baseline["gross_profit_change_usd"] = (
    category_2020_vs_baseline["gross_profit_2020_usd"]
    - category_2020_vs_baseline["baseline_gross_profit_usd"]
)

category_2020_vs_baseline["gross_profit_pct_change"] = np.where(
    category_2020_vs_baseline["baseline_gross_profit_usd"] != 0,
    category_2020_vs_baseline["gross_profit_change_usd"]
    / category_2020_vs_baseline["baseline_gross_profit_usd"],
    np.nan
)

# Margin change.
# This is a percentage point change, not a percent change.
category_2020_vs_baseline["gross_margin_point_change"] = (
    category_2020_vs_baseline["gross_margin_2020_pct"]
    - category_2020_vs_baseline["baseline_gross_margin_pct"]
)

# Sort categories from biggest revenue decline to biggest revenue increase.
category_2020_vs_baseline = category_2020_vs_baseline.sort_values(
    "revenue_change_usd"
)


# ------------------------------------------------------------
# 13. CATEGORY CONTRIBUTION TO TOTAL 2020 CHANGE
# ------------------------------------------------------------
# This answers an executive-friendly question:
#
# "Which categories explain the overall revenue change?"
#
# A category might be down 80%, but if it was tiny, it may not
# matter much to the total business.
#
# Contribution tells us which categories actually moved the company.

total_category_revenue_change = category_2020_vs_baseline["revenue_change_usd"].sum()

category_2020_vs_baseline["share_of_total_revenue_change"] = np.where(
    total_category_revenue_change != 0,
    category_2020_vs_baseline["revenue_change_usd"]
    / total_category_revenue_change,
    np.nan
)

total_category_profit_change = category_2020_vs_baseline["gross_profit_change_usd"].sum()

category_2020_vs_baseline["share_of_total_gross_profit_change"] = np.where(
    total_category_profit_change != 0,
    category_2020_vs_baseline["gross_profit_change_usd"]
    / total_category_profit_change,
    np.nan
)


# ------------------------------------------------------------
# 14. EARLY 2021 DIRECTIONAL CHECK
# ------------------------------------------------------------
# We only have data through 2/20/2021.
#
# That means we cannot compare full-year 2021 against full-year 2020
# or full-year historical performance.
#
# But we can compare Jan 1–Feb 20, 2021 against the same date window
# from prior years.
#
# This is a directional check only.
# It should not be framed as proof of recovery or non-recovery.

# Identify the month/day from the order date.
df["month_day"] = df["order_date"].dt.strftime("%m-%d")

# Keep Jan 1 through Feb 20 for all years.
jan_feb_20_df = df[df["month_day"] <= "02-20"].copy()

# Create an order-level version for the Jan-Feb 20 window.
jan_feb_20_order_summary = (
    jan_feb_20_df.groupby(
        ["order_number", "customerkey", "order_date", "order_year", "analysis_period"],
        as_index=False
    )
    .agg(
        order_revenue_usd=("total_revenue_usd", "sum"),
        order_gross_profit_usd=("total_gross_profit_usd", "sum"),
        order_quantity=("quantity", "sum")
    )
)

early_year_performance = (
    jan_feb_20_order_summary.groupby(
        ["order_year", "analysis_period"],
        as_index=False
    )
    .agg(
        revenue_usd=("order_revenue_usd", "sum"),
        gross_profit_usd=("order_gross_profit_usd", "sum"),
        customers=("customerkey", "nunique"),
        orders=("order_number", "nunique"),
        quantity_sold=("order_quantity", "sum")
    )
)

early_year_performance["avg_order_value_usd"] = np.where(
    early_year_performance["orders"] > 0,
    early_year_performance["revenue_usd"] / early_year_performance["orders"],
    np.nan
)

early_year_performance["orders_per_customer"] = np.where(
    early_year_performance["customers"] > 0,
    early_year_performance["orders"] / early_year_performance["customers"],
    np.nan
)

early_year_performance["revenue_per_customer_usd"] = np.where(
    early_year_performance["customers"] > 0,
    early_year_performance["revenue_usd"] / early_year_performance["customers"],
    np.nan
)

early_year_performance["gross_margin_pct"] = np.where(
    early_year_performance["revenue_usd"] > 0,
    early_year_performance["gross_profit_usd"] / early_year_performance["revenue_usd"],
    np.nan
)

# Build historical Jan 1-Feb 20 baseline from 2016-2019.
early_2021_baseline = (
    early_year_performance[early_year_performance["order_year"].isin(baseline_years)]
    .agg(
        revenue_usd=("revenue_usd", "mean"),
        gross_profit_usd=("gross_profit_usd", "mean"),
        customers=("customers", "mean"),
        orders=("orders", "mean"),
        quantity_sold=("quantity_sold", "mean"),
        avg_order_value_usd=("avg_order_value_usd", "mean"),
        orders_per_customer=("orders_per_customer", "mean"),
        revenue_per_customer_usd=("revenue_per_customer_usd", "mean"),
        gross_margin_pct=("gross_margin_pct", "mean")
    )
    .reset_index()
    .rename(columns={"index": "metric", 0: "historical_jan_feb20_baseline"})
)

early_2021_actual = (
    early_year_performance[early_year_performance["order_year"] == 2021]
    .agg(
        revenue_usd=("revenue_usd", "sum"),
        gross_profit_usd=("gross_profit_usd", "sum"),
        customers=("customers", "sum"),
        orders=("orders", "sum"),
        quantity_sold=("quantity_sold", "sum"),
        avg_order_value_usd=("avg_order_value_usd", "mean"),
        orders_per_customer=("orders_per_customer", "mean"),
        revenue_per_customer_usd=("revenue_per_customer_usd", "mean"),
        gross_margin_pct=("gross_margin_pct", "mean")
    )
    .reset_index()
    .rename(columns={"index": "metric", 0: "early_2021_value"})
)

early_2021_directional_check = early_2021_baseline.merge(
    early_2021_actual,
    on="metric",
    how="left"
)

early_2021_directional_check["absolute_change"] = (
    early_2021_directional_check["early_2021_value"]
    - early_2021_directional_check["historical_jan_feb20_baseline"]
)

early_2021_directional_check["pct_change"] = np.where(
    early_2021_directional_check["historical_jan_feb20_baseline"] != 0,
    early_2021_directional_check["absolute_change"]
    / early_2021_directional_check["historical_jan_feb20_baseline"],
    np.nan
)


# ------------------------------------------------------------
# 15. OPTIONAL DELIVERY PERFORMANCE CHECK
# ------------------------------------------------------------
# This is included because delivery date exists in the data.
#
# This is NOT meant to prove causality.
# It simply checks whether delivery performance looks different
# in 2020 compared to the historical baseline.
#
# If delivery delays increased, this may become a deeper Phase 2
# or Phase 3 investigation.

df["delivery_days"] = (df["delivery_date"] - df["order_date"]).dt.days

# Simple delivery status logic.
# This is just a placeholder.
# In a real client project, we would replace "7 days" with the
# company's actual delivery promise or SLA.
df["delivery_status"] = np.select(
    [
        df["delivery_date"].isna(),
        df["delivery_days"] <= 7,
        df["delivery_days"] > 7
    ],
    [
        "Unknown",
        "On-Time",
        "Late"
    ],
    default="Unknown"
)

delivery_annual_summary = (
    df.groupby(["order_year", "analysis_period", "delivery_status"], as_index=False)
    .agg(
        revenue_usd=("total_revenue_usd", "sum"),
        gross_profit_usd=("total_gross_profit_usd", "sum"),
        customers=("customerkey", "nunique"),
        orders=("order_number", "nunique")
    )
)

delivery_totals_by_year = (
    delivery_annual_summary.groupby("order_year", as_index=False)
    .agg(total_orders=("orders", "sum"))
)

delivery_annual_summary = delivery_annual_summary.merge(
    delivery_totals_by_year,
    on="order_year",
    how="left"
)

delivery_annual_summary["order_share"] = np.where(
    delivery_annual_summary["total_orders"] > 0,
    delivery_annual_summary["orders"] / delivery_annual_summary["total_orders"],
    np.nan
)


# ------------------------------------------------------------
# 16. FINAL OUTPUT TABLES
# ------------------------------------------------------------
# These are the main outputs from Phase 1.
#
# These tables can feed:
# - notebook analysis
# - executive charts
# - dashboard pages
# - case study visuals
# - recommendation slides

phase_1_outputs = {
    "cleaned_line_item_data": df,
    "order_summary": order_summary,
    "monthly_performance": monthly_performance,
    "annual_performance": annual_performance,
    "baseline_vs_2020": baseline_vs_2020,
    "monthly_2020_vs_baseline": monthly_2020_vs_baseline,
    "category_annual_performance": category_annual_performance,
    "category_mix_by_year": category_mix_by_year,
    "category_2020_vs_baseline": category_2020_vs_baseline,
    "early_2021_directional_check": early_2021_directional_check,
    "delivery_annual_summary": delivery_annual_summary
}


# ------------------------------------------------------------
# 17. QUICK QUALITY CHECKS
# ------------------------------------------------------------
# These are simple review checks to make sure the outputs look sane.
# In a notebook, you can replace print statements with display()
# or just call the dataframe names directly.

print("PHASE 1 COMPLETE")
print("----------------")
print("Tables created:")

for table_name in phase_1_outputs.keys():
    print(f"- {table_name}")

print("\nDate range in cleaned data:")
print(df["order_date"].min(), "to", df["order_date"].max())

print("\nAnnual Performance:")
print(annual_performance)

print("\nHistorical Baseline vs 2020:")
print(baseline_vs_2020)

print("\nMonthly 2020 vs Historical Monthly Baseline:")
print(monthly_2020_vs_baseline[
    [
        "month_name",
        "revenue_2020_usd",
        "baseline_revenue_usd",
        "revenue_change_usd",
        "revenue_pct_change",
        "customers_2020",
        "baseline_customers",
        "customer_pct_change",
        "avg_order_value_2020_usd",
        "baseline_avg_order_value_usd",
        "aov_pct_change"
    ]
])

print("\nCategory 2020 vs Historical Baseline:")
print(category_2020_vs_baseline[
    [
        "category",
        "baseline_revenue_usd",
        "revenue_2020_usd",
        "revenue_change_usd",
        "revenue_pct_change",
        "baseline_gross_profit_usd",
        "gross_profit_2020_usd",
        "gross_profit_change_usd",
        "gross_profit_pct_change"
    ]
])

print("\nEarly 2021 Directional Check:")
print(early_2021_directional_check)
```

```python
# If you do not have an actual predictive model, you can still create a historic proxy using cumulative profit by customer and percentile grouping.
# -----------------------------------
# Build cumulative profit per customer-period
# -----------------------------------

customer_period_activity = customer_period_activity.sort_values(
    ["customer_id", "years_since_joining"]
)

customer_period_activity["cumulative_profit_usd"] = (
    customer_period_activity.groupby("customer_id")["profit_usd"].cumsum()
)

# Bring in final CLV to assign customers to percentile groups
customer_period_with_segment = customer_period_activity.merge(
    customer_summary[["customer_id", "clv_profit_usd"]],
    on="customer_id",
    how="left"
)

# Assign percentile groups using final CLV
customer_period_with_segment["clv_percentile"] = (
    customer_period_with_segment["clv_profit_usd"].rank(pct=True) * 100
)

def assign_segment(p):
    if p <= 10:
        return "10th_percentile"
    elif 45 <= p <= 55:
        return "50th_percentile"
    elif p >= 90:
        return "90th_percentile"
    else:
        return "other"

customer_period_with_segment["customer_segment"] = (
    customer_period_with_segment["clv_percentile"].apply(assign_segment)
)

# Average line
avg_curve = (
    customer_period_with_segment.groupby("years_since_joining")["cumulative_profit_usd"]
    .mean()
    .reset_index()
)
avg_curve["customer_segment"] = "average"
avg_curve = avg_curve.rename(columns={"cumulative_profit_usd": "predicted_clv_profit_usd"})

# Percentile-group curves
segment_curves = (
    customer_period_with_segment.loc[customer_period_with_segment["customer_segment"] != "other"]
    .groupby(["years_since_joining", "customer_segment"])["cumulative_profit_usd"]
    .mean()
    .reset_index()
    .rename(columns={"cumulative_profit_usd": "predicted_clv_profit_usd"})
)

predicted_clv_curves = pd.concat([segment_curves, avg_curve], ignore_index=True)
predicted_clv_curves = predicted_clv_curves.sort_values(["customer_segment", "years_since_joining"])

predicted_clv_curves.head()
```

```python
# -----------------------------------
# Add acquisition month
# -----------------------------------
customer_summary["acquisition_month"] = customer_summary["acquisition_date"].dt.to_period("M").astype(str)

# If you want proxy historic CLV horizons, define lifetime days cutoffs
customer_transactions_with_age = customer_transactions.copy()
customer_transactions_with_age["days_since_first_order"] = (
    customer_transactions_with_age["order_date"] - customer_transactions_with_age["acquisition_date"]
).dt.days

def profit_within_days(days):
    temp = customer_transactions_with_age.loc[
        customer_transactions_with_age["days_since_first_order"] <= days
    ]
    out = (
        temp.groupby("customer_id")["total_gross_profit_usd"]
        .sum()
        .reset_index(name=f"profit_{days}d")
    )
    return out

profit_1yr = profit_within_days(365)
profit_3yr = profit_within_days(365 * 3)
profit_5yr = profit_within_days(365 * 5)
profit_10yr = profit_within_days(365 * 10)

clv_horizons = (
    customer_summary[["customer_id", "acquisition_month", "cac_usd"]]
    .merge(profit_1yr, on="customer_id", how="left")
    .merge(profit_3yr, on="customer_id", how="left")
    .merge(profit_5yr, on="customer_id", how="left")
    .merge(profit_10yr, on="customer_id", how="left")
    .fillna(0)
)

for col in ["profit_365d", "profit_1095d", "profit_1825d", "profit_3650d"]:
    clv_horizons[col] = clv_horizons[col] - clv_horizons["cac_usd"]

clv_vs_cac_by_cohort = (
    clv_horizons.groupby("acquisition_month")
    .agg(
        predicted_total_clv_01yr=("profit_365d", "mean"),
        predicted_total_clv_03yr=("profit_1095d", "mean"),
        predicted_total_clv_05yr=("profit_1825d", "mean"),
        predicted_total_clv_10yr=("profit_3650d", "mean"),
        cac_usd=("cac_usd", "mean")
    )
    .reset_index()
    .sort_values("acquisition_month")
)

clv_vs_cac_by_cohort.head()
```

```python
# Revenue within 5 years of acquisition
revenue_5yr = (
    customer_transactions.loc[customer_transactions["days_since_first_order"] <= 365 * 5]
      .groupby("customer_id")["total_revenue_usd"]
      .sum()
      .reset_index(name="total_revenue_5yr")
)

customer_5yr_revenue = (
    customer_summary[["customer_id", "acquisition_date"]]
    .merge(revenue_5yr, on="customer_id", how="left")
    .fillna({"total_revenue_5yr": 0})
)

customer_5yr_revenue["acquisition_month"] = (
    pd.to_datetime(customer_5yr_revenue["acquisition_date"]).dt.to_period("M").astype(str)
)

mean_median_5yr_clv = (
    customer_5yr_revenue.groupby("acquisition_month")["total_revenue_5yr"]
    .agg(
        mean_5yr_total_clv="mean",
        median_5yr_total_clv="median"
    )
    .reset_index()
    .sort_values("acquisition_month")
)

mean_median_5yr_clv.head()
```

```python
# Customer-level lifetime revenue
ltr_dist_source = customer_summary[["customer_id", "total_revenue_usd"]].copy()

# Create bins (adjust as needed)
bins = [0, 25, 50, 75, 100, 150, 200, 300, 500, 1000, np.inf]
labels = ["0-25", "25-50", "50-75", "75-100", "100-150", "150-200", "200-300", "300-500", "500-1000", "1000+"]

ltr_dist_source["lifetime_revenue_bin"] = pd.cut(
    ltr_dist_source["total_revenue_usd"],
    bins=bins,
    labels=labels,
    right=False
)

ltr_distribution = (
    ltr_dist_source.groupby("lifetime_revenue_bin", observed=False)
    .agg(customer_count=("customer_id", "nunique"))
    .reset_index()
)

ltr_distribution["customer_pct"] = (
    ltr_distribution["customer_count"] / ltr_distribution["customer_count"].sum() * 100
)

ltr_distribution["cumulative_pct_customers"] = ltr_distribution["customer_pct"].cumsum()

ltr_distribution.head()
```

```python
concentration_source = customer_summary[["customer_id", "total_revenue_usd"]].copy()

# If you want 5-year LTV instead of lifetime, replace total_revenue_usd with total_revenue_5yr
concentration_source["ltv_5yr_usd"] = concentration_source["total_revenue_usd"]  # placeholder

# Sort highest to lowest for "top x% customers account for y% value"
concentration_source = concentration_source.sort_values(
    ["ltv_5yr_usd", "total_revenue_usd"], ascending=False
).reset_index(drop=True)

concentration_source["customer_rank"] = np.arange(1, len(concentration_source) + 1)
concentration_source["customer_pct_ranked"] = (
    concentration_source["customer_rank"] / len(concentration_source) * 100
)

concentration_source["cumulative_ltv_pct"] = (
    concentration_source["ltv_5yr_usd"].cumsum() / concentration_source["ltv_5yr_usd"].sum() * 100
)

concentration_source["cumulative_revenue_pct"] = (
    concentration_source["total_revenue_usd"].cumsum() / concentration_source["total_revenue_usd"].sum() * 100
)

concentration_source["perfect_equality_pct"] = concentration_source["customer_pct_ranked"]

sales_customer_concentration = concentration_source[[
    "customer_pct_ranked",
    "cumulative_ltv_pct",
    "cumulative_revenue_pct",
    "perfect_equality_pct"
]]

sales_customer_concentration.head()
```

```python
# Example: last 12 months of revenue per customer
max_order_date = customer_transactions["order_date"].max()
start_date = max_order_date - pd.DateOffset(months=12)

recent_revenue = (
    customer_transactions.loc[customer_transactions["order_date"] >= start_date]
      .groupby("customer_id")["total_revenue_usd"]
      .sum()
      .reset_index(name="revenue_last_12mo_usd")
)

recent_revenue = recent_revenue.sort_values("revenue_last_12mo_usd").reset_index(drop=True)

recent_revenue["cumulative_pct_customers"] = (
    (recent_revenue.index + 1) / len(recent_revenue) * 100
)

recent_revenue["cumulative_pct_revenue"] = (
    recent_revenue["revenue_last_12mo_usd"].cumsum() / recent_revenue["revenue_last_12mo_usd"].sum() * 100
)

sales_customer_distribution = recent_revenue.rename(columns={
    "revenue_last_12mo_usd": "revenue_threshold_usd"
})[[
    "revenue_threshold_usd",
    "cumulative_pct_customers",
    "cumulative_pct_revenue"
]]

sales_customer_distribution.head()
```
